```python
ruta_completa = raiz + '/02_Datos/01_Originales/' + nombre_fichero_datos
print(ruta_completa)
datos = pd.read_csv(ruta_completa,index_col=0)
datos
```

```python
#Solo coger datos con mas de un valor
serie_unico = datos.reset_index().nunique()
columnas_utiles = serie_unico[serie_unico>1].index.tolist()
```

```python
df = datos.reset_index()[columnas_utiles].copy()
```

```python
df.sort_values('DATE')
```

```python
df_train = df.loc[:"2022"]
df_validacion = df.loc["2022":]
```
