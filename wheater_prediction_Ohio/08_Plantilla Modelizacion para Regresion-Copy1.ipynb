```python
from  sklearn.model_selection import TimeSeriesSplit
time_cv = TimeSeriesSplit(5, test_size = 30)
```

```python
train_x.shape, train_y.shape
```

```python
random_search = RandomizedSearchCV(estimator = pipe,
                                   param_distributions = grid, 
                                   n_iter = 25, 
                                   cv = time_cv, 
                                   scoring = 'neg_mean_absolute_error', 
                                   verbose = 0,
                                   n_jobs = -1)




modelo = random_search.fit(train_x,train_y)
```

```python
z =pd.DataFrame(random_search.cv_results_).sort_values(by = 'rank_test_score')
z
```

```python
# 10 variables: mae mean 3 mae std 0.4
modelo.best_estimator_
```

```python
from sklearn.inspection import permutation_importance
```

```python
def ranking_per(predictoras,permutacion):
    ranking_per = pd.DataFrame({'variable': predictoras.columns, 'importancia_per': permutacion.importances_mean}).sort_values(by = 'importancia_per', ascending = False)
    ranking_per['ranking_per'] = np.arange(0,ranking_per.shape[0])
    return(ranking_per)
```

```python
m = modelo.best_estimator_.steps[0][1]
permutacion = permutation_importance(m, 
                                     x, y, 
                                     
                                     n_repeats=5, n_jobs = -1)
```

```python
rank_per = ranking_per(x,permutacion)
rank_per
```
